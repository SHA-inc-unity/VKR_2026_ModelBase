# CatBoost Crypto Forecasting Diagnostics
This notebook visualizes backtest results, anomalies, and confidence scores.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../outputs/backtest_results/backtest_results.csv', parse_dates=['timestamp'])
df = df.sort_values('timestamp')
plt.figure(figsize=(16,6))
plt.plot(df['timestamp'], df['target_direct'], label='True Close')
plt.plot(df['timestamp'], df['direct_pred'], label='Direct Model Pred')
plt.fill_between(df['timestamp'], df['range_pred_low'], df['range_pred_high'], color='orange', alpha=0.2, label='Range Model Corridor')
plt.legend()
plt.title('Direct and Range Model Predictions')
plt.show()
```

```python
# Anomaly visualization
anomalies = df[df['anomaly_flag'] == True]
plt.figure(figsize=(16,4))
plt.plot(df['timestamp'], df['target_direct'], label='True Close')
plt.scatter(anomalies['timestamp'], anomalies['target_direct'], color='red', label='Anomaly', marker='x')
plt.legend()
plt.title('Anomalies in Backtest')
plt.show()
```

```python
# Confidence score
plt.figure(figsize=(16,2))
plt.plot(df['timestamp'], df['confidence'], label='Confidence')
plt.title('Prediction Confidence Over Time')
plt.show()
```
